```python
file_2024 = "DATOS HISTÓRICOS 2024_TODAS ESTACIONES.xlsx"
file_2025 = "DATOS HISTÓRICOS 2025_TODAS ESTACIONES.xlsx"
sheets = ["NO2", "NO3", "NO", "NTE2", "SE"]

output_file = "datos_unificados_estaciones_simple.xlsx"

with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
    for sheet in sheets:
        # --- Leer 2024 con encabezado correcto ---
        df24 = pd.read_excel(file_2024, sheet_name=sheet, header=0)
        cols_24 = df24.columns.tolist()  # guardamos columnas

        # --- Leer 2025, saltando las 2 primeras filas y forzando columnas de 2024 ---
        df25 = pd.read_excel(file_2025, sheet_name=sheet, header=None, skiprows=2)
        # recortamos/ajustamos a las columnas de 2024
        df25 = df25.iloc[:, :len(cols_24)]
        df25.columns = cols_24  # forzamos nombres de 2024

        # --- Concatenar verticalmente ---
        merged = pd.concat([df24, df25], axis=0, ignore_index=True)

        # --- Guardar hoja ---
        merged.to_excel(writer, sheet_name=sheet, index=False)
```

```python
file = "datos_unificados_estaciones_simple.xlsx"

# Leer todas las hojas en un diccionario
xls = pd.ExcelFile(file)
sheets = xls.sheet_names
print("Hojas disponibles:", sheets)

for sheet in sheets:
    print(f"\n--- Hoja: {sheet} ---")
    df = pd.read_excel(file, sheet_name=sheet)
    
    # Nombres de columnas
    print("Columnas:", df.columns.tolist())
    
    # Valores nulos por columna
    nulos = df.isnull().sum()
    print("\nValores nulos por columna:")
    print(nulos)
    
    # Valores cero por columna (solo columnas numéricas)
    ceros = (df.select_dtypes(include='number') == 0).sum()
    print("\nValores cero por columna (numéricas):")
    print(ceros)
```
